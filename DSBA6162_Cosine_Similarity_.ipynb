```python
import numpy as np

data = np.array([
    [1.5, 1.7],
    [2.0, 1.9],
    [1.6, 1.8],
    [1.2, 1.5],
    [1.5, 1.0]
])

print(data)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import sklearn.metrics


x = np.array([[1.4, 1.6]])

cosine_sim_matrix = cosine_similarity(data, x)


display(cosine_sim_matrix)
```
